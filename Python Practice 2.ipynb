### Word Counter

```python
dict = {}
sentence = input('Write a sentence:').lower().split()
for word in sentence:
    if word in dict:
        dict[word] += 1
    else:
        dict[word] = 1
print(dict)
```

### Max Finder

```python
def max_finder(list):
    a=list[0]
    for item in list:
        if item >= a:
            a = item
    return a

print(max_finder([1,103,-19,123,59,32,3200,10000,-100000]))
```

### Common elements

```python
number = input('Type in Done to Finish your Set \nPlease Enter a Number for your Set A:')
set_a = set()
set_b = set()
while number != 'Done':
    set_a.add(int(number))
    number = input('Type in Done to Finish your Set \nPlease Enter a Number for your Set A:')

number2 = input('Type in Done to Finish your Set \nPlease Enter a Number for your Set B:')
while number2 != 'Done':
    set_b.add(int(number2))
    number2 = input('Type in Done to Finish your Set \nPlease Enter a Number for your Set B:')
    
print(f'Set A is {set_a}\nSet B is {set_b}\nThe Common Numbers are {set_a.intersection(set_b)}')
    
```

### Temperature Converter

```python
# function takes string arguments with F or C at end
def converter(temp):
    if temp[-1] == 'F':
        return (5/9)*(float(temp[:-1])-32)
    if temp[-1] == 'C':
        return ((9/5)*float(temp[:-1]))+32
print(converter('63.5F'))
```

### Password Strength Checker

```python
# Password requires, at least, a number, a special character, and a length of 8.
def numchecker(text):
    o=[]
    for char in text:
        o.append(char.isdigit())
    return any(o)
       
def specchecker(text):
    spec_chars = set("@£$%&*!?#")
    o=[]
    for char in text:
        if char in spec_chars:
            o.append(True)
        else:
            o.append(False)
    return any(o)   
```

```python
passed = False
while not passed:
    pword = input('Type your password here:')
    if len(pword) <= 8:
        print('Your password must be at least 8 characters')
    elif not specchecker(pword):
        print('Your password must have at least one special character "@£$%&*!?#"')
    elif not numchecker(pword):
        print('Your password must contain at least one digit')
    else:
        print('Password Accepted!')
        passed = True
```
